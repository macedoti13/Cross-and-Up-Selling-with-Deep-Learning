```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_parquet("../data/raw/puc_vendas.parquet")
print("Total number of transactions: ", df.shape[0])
df.head()
```

```python
df.rename(
    columns={
        "COD_CUPOM": "cod_cupom",
        "COD_CLIENTE": "client_id",
        "CLIENTE_FISICO_JURIDICO": "client_type",
        "SEXO_CLIENTE": "client_gender",
        "DTNASCIMENTO_CLIENTE": "client_birthdate",
        "COD_SKU": "product_id",
        "SKU": "product_name",
        "CATEGORIA_SKU": "product_category",
        "SUBCATEGORIA_SKU": "product_subcategory",
        "COD_LOJA": "store_id",
        "UF_CIDADE": "city_state",
        "COD_CIDADE": "city_id",
        "NOME_CIDADE": "city_name",
        "DATA_CUPOM": "sale_date",
        "UNIDADES": "units",
        "IDENTIFICADOR_PROMOCIONAL": "promotion_id",
        "PRECO_REGULAR": "regular_price",
        "TOTAL_DESCONTO": "total_discount",
        "TOTAL_BRUTO": "total_gross",
        "TOTAL_LIQUIDO": "total_net",   
    },
    inplace=True
)
df.head()
```

```python
print(df.city_state.value_counts(), df.city_id.value_counts(), df.city_name.value_counts())
```

```python
df.head()
```

```python
df["sale"] = df[['client_id', 'store_id', 'sale_date']].astype(str).agg('_'.join, axis=1)
df["sale_id"] = df.groupby('sale').ngroup() + 1
df.drop(columns=["sale"], inplace=True)
df.head()
```

```python
products = df[["product_id", "product_name", "product_category", "product_subcategory", "regular_price"]].drop_duplicates()
print(f"Number of unique products: {products.shape[0]}")
products.head()
```

```python
products.drop_duplicates(subset=["product_name", "product_category", "product_subcategory"], keep="first", inplace=True)
products = products[~products["product_category"].isin(["SERVIÇOS", "DIVERSOS", "MANIPULADOS", "IMOBILIZADO"])]

# get subcategories with less than 10 appearances
under_11_subcategories = products["product_subcategory"].value_counts()
under_11_subcategories = under_11_subcategories[under_11_subcategories < 11]
under_11_subcategories = list(under_11_subcategories.to_frame().reset_index().product_subcategory.values)
under_11_subcategories
products = products[~products["product_subcategory"].isin(under_11_subcategories)]

# get products with price between 1 and 9999
products = products[(products["regular_price"] >= 1) & (products["regular_price"] <= 9999)] 

print(f"Number of unique products: {products.shape[0]}")
```

```python
df = df[df.product_id.isin(products.product_id)]
print("Total number of transactions: ", df.shape[0])
print(f"Number of unique products: {df.product_id.nunique()}")
df.head()
```

```python
clients = df.groupby(
    ["client_id", "client_type", "client_gender", "client_birthdate"]
)[["total_gross", "total_net", "total_discount"]].agg(
    ["sum", "mean", "max", "min"]
).reset_index()
clients.columns = ["_".join(col).strip() for col in clients.columns.values]
clients = clients.reset_index()
clients.head()
```
