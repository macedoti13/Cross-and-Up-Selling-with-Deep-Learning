```python
import pandas as pd
```

```python
input_file = 'data/puc_campanhas.csv'
output_file = 'data/puc_campanhas_deduplicated.csv'
batch_size = 10000
```

```python
# Open the output file in write mode
with open(output_file, 'w') as f_out:
    # Initialize a variable to keep track of whether the header has been written
    header_written = False
    
    # Initialize a counter for processed rows
    total_processed_rows = 0
    
    # Read the file in chunks
    for chunk in pd.read_csv(input_file, chunksize=batch_size):
        # Drop duplicates within the chunk
        chunk = chunk.drop_duplicates()
        
        # Write the chunk to the output file
        if not header_written:
            # Write the header if it hasn't been written yet
            chunk.to_csv(f_out, index=False, mode='a', header=True)
            header_written = True
        else:
            # Append without writing the header
            chunk.to_csv(f_out, index=False, mode='a', header=False)
        
        # Update and print the number of rows processed
        total_processed_rows += len(chunk)
        print(f"Processed {total_processed_rows} rows so far.")

print("Finished processing all chunks.")
```
